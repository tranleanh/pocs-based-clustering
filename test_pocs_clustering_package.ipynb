### Import

```python
import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from pocs_cluster_analysis import pocs_clustering

np.set_printoptions(suppress=True)
```

### Demo

```python
num_clusters = 10
X, y = make_blobs(n_samples=5000, centers=num_clusters, cluster_std=0.5, random_state=0)

plt.figure(figsize=(8,8))
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.show()
```

```python
# POSC-based Clustering Demo
start_time = time.time()
centroids, labels = pocs_clustering(X, num_clusters, 100)
end_time = time.time()
proc_time = end_time - start_time
print('Execution time:', proc_time)

# Display Results
plt.figure(figsize=(8,8))
plt.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='red')
plt.title('Clustering Results')
plt.show()
```
